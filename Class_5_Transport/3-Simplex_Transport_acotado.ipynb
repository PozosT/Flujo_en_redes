## **Lectura 3: Símplex de transporte con variables acotadas**

Un **problema de transporte con variables acotadas** extiende el modelo estándar permitiendo cotas inferiores y superiores sobre cada variable de flujo $x_{ij}$. La formulación completa es:

$$\min \sum_{i=1}^{m} \sum_{j=1}^{n} c_{ij}\, x_{ij}$$

$$\text{s.a.} \quad \sum_{j=1}^{n} x_{ij} = a_i \qquad i = 1, \ldots, m$$

$$\sum_{i=1}^{m} x_{ij} = b_j \qquad j = 1, \ldots, n$$

$$l_{ij} \leq x_{ij} \leq u_{ij} \qquad \forall\, i, j$$

donde $a_i$ es la oferta del origen $i$, $b_j$ es la demanda del destino $j$, $l_{ij} \geq 0$ es la cota inferior y $u_{ij}$ es la capacidad del arco $(i,j)$.

Este tipo de modelo aparece, por ejemplo, en problemas de transporte con contratos mínimos garantizados ($l_{ij} > 0$) o con capacidades de ruta ($u_{ij} < \infty$).

### Revisión del método símplex de transporte estándar

El método símplex de transporte sin cotas opera de la siguiente manera:

1. Se construye una solución básica factible inicial (SBFI) con $m + n - 1$ variables básicas.
2. Se calculan los potenciales $u_i$ y $v_j$ a partir de las condiciones $u_i + v_j = c_{ij}$ para los arcos básicos.
3. Se computan los costos reducidos $\bar{c}_{ij} = c_{ij} - u_i - v_j$ para las variables no básicas.
4. Si todos los costos reducidos son no negativos, la SBFI es **óptima**.
5. Si existe algún $\bar{c}_{ij} < 0$, se selecciona esa variable como entrante, se identifica el ciclo y se realiza el pivote.

La variante acotada extiende este procedimiento para manejar las cotas $l_{ij}$ y $u_{ij}$ de forma eficiente, sin aumentar el tamaño del problema.

El problema con cotas puede abordarse de dos maneras:

1. **Transformación a forma estándar:** Fijar $x_{ij} = y_{ij} + l_{ij}$ elimina la cota inferior mediante un cambio de variable, ajustando ofertas y demandas. Para las cotas superiores se necesitan variables de holgura adicionales ($s_{ij} = u_{ij} - x_{ij} \geq 0$), lo que aumenta el tamaño del problema.

2. **Símplex acotado (opción eficiente):** Modificar directamente el método símplex para que las variables no básicas puedan encontrarse tanto en su cota inferior como en su cota superior, sin agregar variables ni restricciones. Esta es la alternativa que se desarrolla en esta lectura.

### Paso 1: Clasificación de variables y cálculo de potenciales

En el símplex de transporte acotado, cada variable puede estar en uno de tres estados:

- **Variables básicas:** $l_{ij} < x_{ij} < u_{ij}$. Sus costos reducidos son cero: $\bar{c}_{ij} = c_{ij} - u_i - v_j = 0$.
- **Variables no básicas en cota inferior:** $x_{ij} = l_{ij}$. La variable no se puede disminuir; solo puede aumentar.
- **Variables no básicas en cota superior:** $x_{ij} = u_{ij}$. La variable no se puede aumentar; solo puede disminuir.

Los potenciales $u_i$ y $v_j$ se calculan igual que en el caso sin cotas: fijando $u_1 = 0$ y resolviendo $u_i + v_j = c_{ij}$ para todos los arcos básicos.

**Nota sobre soluciones degeneradas:** si una variable básica coincide exactamente con su cota inferior o superior ($x_{ij} = l_{ij}$ o $x_{ij} = u_{ij}$), la solución es **degenerada acotada**.

### Paso 2: Condiciones de optimalidad

La solución actual es óptima si y solo si, para todas las variables no básicas:

1. Si $(i,j)$ es **básico**: $\bar{c}_{ij} = c_{ij} - u_i - v_j = 0$.

2. Si $(i,j)$ es **no básico en cota inferior** ($x_{ij} = l_{ij}$): no conviene aumentar $x_{ij}$, luego se requiere $\bar{c}_{ij} \geq 0$.

3. Si $(i,j)$ es **no básico en cota superior** ($x_{ij} = u_{ij}$): no conviene disminuir $x_{ij}$, luego se requiere $\bar{c}_{ij} \leq 0$.

Una violación de estas condiciones indica la variable que debe entrar a la base.

### Paso 3: Selección de la variable entrante

Se revisan las variables no básicas en busca de violaciones a la optimalidad:

- **Variables no básicas en cota inferior** ($x_{ij} = l_{ij}$): si $\bar{c}_{ij} < 0$, conviene **aumentar** $x_{ij}$ (entra a la base moviéndose desde $l_{ij}$).
- **Variables no básicas en cota superior** ($x_{ij} = u_{ij}$): si $\bar{c}_{ij} > 0$, conviene **disminuir** $x_{ij}$ (entra a la base moviéndose desde $u_{ij}$).

Entre todas las variables candidatas se selecciona la de mayor violación (mayor $|\bar{c}_{ij}|$).

### Paso 4: Cálculo del incremento $\Delta$ y pivote

Una vez identificada la variable entrante $(p,q)$ y el ciclo $C$ en la tabla de transporte, el incremento máximo $\Delta$ está limitado por la primera variable que alcanza una de sus cotas:

- Si $(p,q)$ entra desde su cota **inferior** ($x_{pq} = l_{pq}$):
  - Los arcos del ciclo en sentido directo aumentan su flujo; el límite superior es $\Delta \leq u_{ij} - x_{ij}$.
  - Los arcos del ciclo en sentido inverso disminuyen su flujo; el límite es $\Delta \leq x_{ij} - l_{ij}$.
  - Se toma $\Delta = \min$ de todos los límites anteriores.
- Si $(p,q)$ entra desde su cota **superior** ($x_{pq} = u_{pq}$): la dirección del recorrido del ciclo se invierte.

El arco que alcanza primero su cota (inferior o superior) es la **variable saliente**. Si el mínimo se alcanza en el propio arco entrante (sin recorrer el ciclo completo), el árbol no cambia y solo se invierte el estado de $(p,q)$ entre $l_{pq}$ y $u_{pq}$.

### Construcción de una solución básica factible inicial con cotas

Para obtener la SBFI con cotas se siguen estos pasos:

**Paso 1:** Fijar todas las variables en su cota inferior: $x_{ij} = l_{ij}$.

![](../Images/TACO6.png)

**Paso 2:** Ajustar ofertas y demandas restando los flujos ya fijados: $a_i' = a_i - \sum_j l_{ij}$ y $b_j' = b_j - \sum_i l_{ij}$.

![](../Images/TACO7.png)

**Paso 3:** Resolver el problema de transporte estándar (sin cotas) con las ofertas y demandas ajustadas $a_i'$ y $b_j'$, usando el método de costo mínimo u otro heurístico para obtener una buena solución inicial.
