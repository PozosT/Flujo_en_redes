## **Lectura 2: Símplex de redes — Estructura triangular y cambios de base**


### Estructura triangular de la base

Una propiedad fundamental del símplex de redes es que la submatriz de base $B$ —formada por las columnas de la matriz de incidencia nodo-arco correspondientes a los arcos del árbol— siempre puede reordenarse como una **matriz triangular**. Esto es consecuencia directa de la estructura de árbol: en todo árbol existe al menos un nodo hoja (nodo de grado 1), cuya fila en $B$ tiene un único elemento no nulo. Eliminando ese nodo y el arco que lo conecta, el árbol resultante conserva la propiedad, por lo que el proceso puede repetirse hasta cubrir todos los nodos.

La estructura triangular tiene dos consecuencias computacionales importantes:

1. El sistema $Bx_B = b$ puede resolverse en $O(n)$ operaciones mediante **sustitución hacia adelante o hacia atrás**, sin necesidad de invertir $B$ explícitamente.
2. La actualización de potenciales tras un cambio de árbol también es $O(n)$: basta recorrer el subárbol afectado por el intercambio.

2. Cualquier árbol generador de la red admite este tipo de reordenamiento, por lo que la resolución del sistema $Bx_B = b$ en el símplex de redes nunca requiere una factorización de costo $O(n^3)$.


### Cambio de base: intercambio de arcos en el árbol

Cuando el símplex de redes determina que un arco $(p,q) \notin T$ tiene costo reducido negativo, se realiza un **pivote**: se agrega $(p,q)$ al árbol, se identifica el ciclo $C$ que se forma, y se saca del árbol el arco que primero alcanza flujo cero al aumentar el flujo en $\Delta$ unidades a lo largo del ciclo.

El parámetro $\Delta$ (denominado $\theta$ en algunas referencias) es el **incremento máximo de flujo** que puede introducirse en el arco entrante sin violar la no negatividad de los flujos:

$$\Delta = \min_{(i,j) \in C^-} x_{ij}$$

donde $C^-$ denota el conjunto de arcos del ciclo $C$ que deben **reducir** su flujo (arcos en sentido contrario al arco entrante). El arco $(k,l) \in C^-$ que alcanza el mínimo es el **arco saliente**: sale de la base con flujo cero.

### Reglas de pivote

La selección del arco entrante (paso 4 del algoritmo) puede realizarse con diferentes **reglas de pivote**, con implicaciones en la eficiencia y en la prevención del ciclado:

- **Regla de Dantzig (mayor costo reducido negativo):** se selecciona el arco con $\bar{c}_{ij}$ más negativo. Tiende a reducir el número de iteraciones, pero no garantiza la terminación en presencia de degeneración.
- **Regla de Bland:** se selecciona siempre el arco de menor índice con costo reducido negativo. Garantiza la terminación (previene el ciclado), aunque puede requerir más iteraciones.
- **Selección aleatoria:** en implementaciones modernas se emplean heurísticas o estrategias de tipo "first eligible arc" para equilibrar calidad de pivote y costo por iteración.
